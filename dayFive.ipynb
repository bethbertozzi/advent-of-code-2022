```python
class Stack:
 
    # create empty list
    def __init__(self):
        self.Elements = []
         
    # push() for insert an element
    def push(self, value):
        self.Elements.append(value)
       
    # pop() for remove an element
    def pop(self):
        return self.Elements.pop()
     
    # empty() check the stack is empty of not
    def empty(self):
        return self.Elements == []
     
    # show() display stack
    def show(self):
        for value in reversed(self.Elements):
            print(value)
 
# Insert_Bottom() insert value at bottom
def BottomInsert(s, value):
   
    # check the stack is empty or not
    if s.empty():
         
        # if stack is empty then call
        # push() method.
        s.push(value)
         
    # if stack is not empty then execute
    # else block
    else:
        popped = s.pop()
        BottomInsert(s, value)
        s.push(popped)
 
# Reverse() reverse the stack
def Reverse(s):
    if s.empty():
        pass
    else:
        popped = s.pop()
        Reverse(s)
        BottomInsert(s, popped)
```

```python
from re import I
import os
from queue import LifoQueue
filename = "day_five_data.txt"
big_data = "aoc_2022_day05_large_input.txt"

def print_list_of_stacks(l):
  for s in l:
    print('NEW STACK')
    s.show()
  print('\n')

def build_stacks(raw_stacks):
  a = []
  rows = raw_stacks.split("\n")
  n = 4
  for i in range(0, len(rows[0]), n):
    s = Stack()
    s.push(rows[0][i:i+n])
    a.append(s)

  for row_index in range(1, len(rows)):
    for i in range(0, len(rows[row_index]), n):
      a[int(i/4)].push(rows[row_index][i:i+n])

  for stack in a:
    stack.pop()
    Reverse(stack)
  return a

def clean_stacks(a):
  for q in a:
    popped = q.pop()
    while(popped == '    ' or popped == '   '):
      popped = q.pop()
    q.push(popped)
  return a

def execute_instr(instr, a):
  txt = instr.split(" ")
  amount = int(txt[1])
  source_index = int(txt[3]) -1
  dest_index = int(txt[5])-1

  popped_q = Stack()
  for i in range(0, amount):
    popped_q.push(a[source_index].pop())
  for i in range(0, amount):
    a[dest_index].push(popped_q.pop())
  
  

raw_stacks, raw_instr = open(filename).read().split('\n\n')
a = build_stacks(raw_stacks)
clean_stacks(a)

for instr in raw_instr.splitlines():
  execute_instr(instr, a)

print_list_of_stacks(a)

```
